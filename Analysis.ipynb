# CREATING DATAFRAME

```python
data = {'party_name': [1,  5, 18,  8,  4,  3, 11, 10,  2,  7,  6,  9 , 12],
       'q8a_1': [751,123, 1, 0, 38, 1, 0 , 0, 0, 0, 0, 0, 0],
       'q9b': [116, 449, 4, 9, 160, 2, 1, 1, 1, 0, 0, 2, 1],
       'q11': [242, 552, 24, 11, 67, 5, 1, 3, 3, 1, 3, 2, 0],
       'q12': [211, 590, 24, 0, 82, 2, 0, 0, 1, 0, 0, 0, 1 ]}
```

```python
df1 = pd.DataFrame(data)
```

```python
df1
```

```python
df1['2012'] = df1['q8a_1'].div(df1['q8a_1'].sum())
```

```python
df1['2014'] = df1['q9b'].div(df1['q9b'].sum())
```

```python
df1['2017'] = df1['q11'].div(df1['q11'].sum())
```

```python
df1['2019'] = df1['q12'].div(df1['q12'].sum())
```

```python
df1
```

```python
df1['expected'] = ((df1['q8a_1'] * df1['2012']) + (df1['q9b'] * df1['2014']) + (df1['q11'] * df1['2017']) + (df1['q12'] * df1['2019'])).div(df1['q8a_1'] + df1['q9b'] + df1['q11'] + df1['q12'])
```

```python
df1
```

# CALCULATING EXPECTED VOTE SHARE

```python
df1['vote_share'] = df1['expected'] * 100
```
